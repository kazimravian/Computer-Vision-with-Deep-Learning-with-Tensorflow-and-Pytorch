<h1><font color = 'brown' size = '6'>
<b>
1.2 How to Load an Image with Keras
</b>
</font>
</h1>

<h1><font color = 'brown green' size = '3'>
<b>
Keras provides the <i>load_img()</i> function for loading an image from file as a PIL image object. The example below loads the Bondi Beach photograph from file as a PIL image and reports details about the loaded image.
</b>
</font></h1>

```python
# example of loading an image with the Keras API
from tensorflow.keras.preprocessing.image import load_img
import matplotlib.pyplot as plt

# load the image
img = load_img('/content/drive/MyDrive/Deep Learning and Computer Vision/images/bondi_beach.jpg')

# report details about image
print(type(img))
print(img.mode)
print(img.format)
print(img.size)
print()

# show the image
plt.imshow(img)
plt.title('Test Image')
```

<h1><font color = 'brown' size = '6'>
<b>
1.3 How to Convert an Image With Keras 
</b>
</font>
</h1>

<h1><font color = 'brown green' size = '3'>
<b>
Keras provides the <i>img_to_array()</i> function for converting a loaded image in PIL format into a NumPy array for use with deep learning models. The API also provides the <i>array_to_img()</i> function that can be used for converting a NumPy array of pixel data into a PIL image. This can be useful if the pixel data is modified in array format because it can be saved or viewed. The example below loads the test image, converts it to a NumPy array, and then converts it back into a PIL image.
</b>
</font></h1>

```python
# example of converting an image with the Keras API
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.preprocessing.image import array_to_img

# load the image
img = load_img('/content/drive/MyDrive/Deep Learning and Computer Vision/images/bondi_beach.jpg')
print(type(img))

# convert image into numpy array
img_array = img_to_array(img)
print(img_array.dtype)
print(img_array.shape)

# convert back to image
img_pil = array_to_img(img_array)
print(type(img_pil))
```

<h1><font color = 'brown' size = '6'>
<b>
1.4 How to Save an Image With Keras 
</b>
</font>
</h1>

<h1><font color = 'brown green' size = '3'>
<b>
The Keras API also provides the save <i>img()</i> function to save an image to file. The function takes the path to save the image, and the image data in NumPy array format.
</b>
</font></h1>

```python
# example of saving an image in keras
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing.image import save_img
from tensorflow.keras.preprocessing.image import img_to_array
import matplotlib.pylab as plt

# load image as grayscale
img = load_img('/content/drive/MyDrive/Deep Learning and Computer Vision/images/bondi_beach.jpg', color_mode = 'grayscale')

# convert img into numpy array
img_array = img_to_array(img)

# save the image as new grayscale file
save_img('/content/drive/MyDrive/Deep Learning and Computer Vision/images/bondi_beach_grayscale.jpg', img_array)

# load image to check for correctly saved
img_save = load_img('/content/drive/MyDrive/Deep Learning and Computer Vision/images/bondi_beach_grayscale.jpg')

print(type(img_save))
print(img_save.format)
print(img_save.mode)
print(img_save.size)
plt.imshow(img_save)
```
