<h1>
<font color = 'brown' size = 7>
<b> Computer Vision </b>
</font>
</h1>

<h1>
<uL>
<font color = 'green brown' size = '5'>

<li>
<b>
Computer vision is a field of study focused on the problem of helping computers to see.
</b>
</li>

<li>
<b>
At an abstract level, the goal of computer vision problems is to use the observed image data to infer something about the world.
</b>
</li>

<li>
<b>
Computer vision is the automated extraction of information from images. Information can mean anything from 3D models, camera position, object detection and recognition to grouping and searching image content.
</b>
</li>

</font>
</ul>
</h1>

<h1>
<font color = 'brown' size = 7>
<b> Computer Vision and Image Processing </b>
</font>
</h1>

<h1>
<uL>
<font color = 'green brown' size = '5'>

<li>
<b>
Computer vision is distinct from image processing.
</b>
</li>

<li>
<b>
Image processing is the process of creating a new image from an existing image, typically simplifying or enhancing the content in some way. It is a type of digital signal processing and is not concerned with understanding the content of
an image.
</b>
</li>

<li>
<b>
A given computer vision system may require image processing to be applied to raw
input, e.g. pre-processing images. Examples of image processing include:

<ul>
<b>
<li>Normalizing photometric properties of the image, such as brightness or color.</li>
</b>
<b>
<li>Cropping the bounds of the image, such as centering an object in a photograph.</li>
</b>
<b>
<li>Removing digital noise from an image, such as digital artifacts from low light levels.</li>
</b>
</ul>

</b>
</li>

</font>
</ul>
</h1>

<h1>
<font color = 'brown' size = 7>
<b>Challenge of Computer Vision</b>
</font>
</h1>

<h1>
<uL>
<font color = 'green brown' size = '5'>

<li>
<b>
The goal of computer vision is to extract useful information from images. This has proved a surprisingly challenging task.
</b>
</li>

<li>
<b>
It has occupied thousands of intelligent and creative minds over the last four decades.
</b>
</li>

<li>
<b>
Despite this we are still far from being able to build a general-purpose “seeing machine.”
</b>
</li>

<li>
<b>
Another reason why it is such a challenging problem is because of the complexity inherent in the visual world.
</b>
</li>

<li>
<b>
A given object may be seen from any orientation, in any lighting conditions,
with any type of occlusion from other objects, and so on.
</b>
</li>

<li>
<b>
A true vision system must be able to see in any of an infinite number of scenes and still extract something meaningful.
</b>
</li>

<li>
<b>
Computers work well for tightly constrained problems, not open unbounded problems like visual perception.
</b>
</li>


</font>
</ul>
</h1>

<h1>
<font color = 'brown' size = 7>
<b>Tasks in Computer Vision</b>
</font>
</h1>

<h1>
<uL>
<font color = 'green brown' size = '5'>
<li>
<b>
The different tasks which computer vision has completed in real world are as under:

<uL>
<font color = 'green brown' size = '5'>
<li>Optical character recognition (OCR)</li>
<li>Machine inspection</li>
<li>Retail (e.g. automated checkouts)</li>
<li>3D model building (photogrammetry)</li>
<li>Medical imaging</li>
<li>Automotive safety</li>
<li>Match move (e.g. merging CGI with live actors in movies)</li>
<li>Motion capture (mocap)</li>
<li>Surveillance</li>
<li>Fingerprint recognition and biometrics</li>
<li>Object Classification: What broad category of object is in the photograph?</li>
<li>Object Identification: Which type of a given object is in this photograph?</li>
<li>Object Verification: Is the object in the photograph?</li>
<li>Object Detection: Where are the objects in the photograph?</li>
<li>Object Landmark Detection: What are the key points for the object in the photograph?</li>
<li>Object Segmentation: What pixels belong to the object in the image?</li>
<li>Object Recognition: What objects are in this photograph and where are they?</li>
</font>
</ul>


</b>
</li>
</font>
</ul>
<font color = 'green brown' size = '5'><b>Other common examples are related to information retrieval; for example: finding new images like an existing image or images.</b></font>

<h1>
<font color = 'brown' size = 7>
<b>Use of Deep Learning in
Computer Vision</b>
</font>
</h1>

<h1>
<uL>
<font color = 'green brown' size = '5'>
<b>Deep Learning Algorithm are used to do various tasks in computer vision such as:</b><br>
<li>
<b>
Automatic Feature Extraction:

<uL>
<li>
A major focus of study in the field of computer vision is on techniques to detect and extract features from digital images.
</li><br>
<li>
Extracted features provide the context for inference about an image, and often the richer the features, the better the inference.
</li><br>
<li>
The promise of deep learning is that complex and useful features can be automatically learned directly from large image datasets.
</li><br>
<li>
More specifically, that a deep hierarchy of rich features can be learned and automatically extracted from images, provided by the multiple deep layers of neural network models.
</li><br>
</ul> 

</b>
</li>
</font>
</ul>
</h1>

<h1>
<uL>
<font color = 'green brown' size = '5'>
<li>
<b>
End-to-End Models:

<ul>
<li>
The deep neural networks allow a single model to subsume two or more traditional models, such as feature extraction and classification.
</li><br>
<li>
It is common to use a single model trained directly on raw pixel values for image classification, and there has been a trend toward replacing pipelines that use a deep neural network model with a single model trained end-to-end directly.
</li><br>
<li>
A good example of this is in object detection and face recognition where initially superior performance was achieved using a deep convolutional neural network for feature extraction.
</li>
</ul>

</b>
</li>
</font>
</ul>
</h1>

<h1>
<uL>
<font color = 'green brown' size = '5'>
<li>
<b>
Model Reuse:

<ul>
<li>
Deep neural networks are typically trained on datasets that are much larger than traditional datasets, e.g. millions or billions of images.
</li><br>
<li>
This allows the models to learn features and hierarchies of features that are general across photographs, which is itself remarkable.
</li><br>
</ul>

</b>
</li>
</font>
</ul>
</h1>

<h1>
<uL>
<font color = 'green brown' size = '5'>
<li>
<b>
Superior Performance:

<ul>
<li>
An important promise of deep neural networks in computer vision is better performance.
</li><br>
<li>It is the dramatically better performance with deep neural networks that has been a catalyst for the growth and interest in the field of deep learning.
</li><br>
<li>
Although the techniques have been around for decades, the spark was the outstanding performance by Alex Krizhevsky, et al. in 2012 for image classification.
</li><br>
<li>
This trend of improvement has continued year-over-year on a range of computer vision tasks.
</li><br>
<li>
We can observe significant performance (mean average precision) improvement since deep learning entered the scene in 2012.
</li><br>
</ul>

</b>
</li>
</font>
</ul>
</h1>

<h1>
<uL>
<font color = 'green brown' size = '5'>
<li>
<b>
General Method:

<ul>
<li>
Perhaps the most important promise of deep learning is that the top-performing models are all developed from the same basic components.
</li><br>
<li>
The impressive results have come from one type of network, called the convolutional neural network (CNN), comprised of convolutional and pooling layers.
</li><br>
<li>
CNN was specifically designed for image data and can be trained on pixel data directly.
</li><br>
</ul>

</b>
</li>
</font>
</ul>
</h1>

<h1>
<font color = 'brown' size = 6>
<b>
Deep Learning Models Used in Computer Vision 
</b>
</font>
</h1>

<h1>
<uL>
<font color = 'green brown' size = '5'>

<li>
<b>
Deep Learning is a large field of study, and not all of it is relevant to computer vision.
</b>
</li><br>

<li>
<b>
From a high-level, there is one method from deep learning that deserves the
most attention for application in computer vision, It is:
<ul> 
<li>
Convolutional Neural Networks (CNNs).
</li>
</ul>
</b>
</li><br>

<li>
<b>
The reason that CNNs are the focus of attention for deep learning models is that they were specifically designed for image data.
</b>
</li><br>

<li>
<b>
Additionally, both of the following network types may be useful for interpreting or developing inference models from the features learned and extracted by CNNs; they are:

<ul> 
<li>
Multilayer Perceptrons (MLP).
</li>
</ul>
<ul> 
<li>
Recurrent Neural Networks (RNNs).
</li>
</ul>
</b>
</li><br>

<b>
<li>
The MLP or fully-connected type neural network layers are useful for developing models that make predictions given the learned features extracted by CNNs.
</li><br>
</b>

<b>
<li>
RNNs, such as LSTMs, may be helpful when working with sequences of images over time, such as with video.
</li>
</b>

</font>
</ul>
